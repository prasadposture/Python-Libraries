# Displot
It is stands for distribution plot. It comes with histplot, kdeplot, ecdfplot and rugplot. It also gives access to seaborn's facetgrid.

### Importing library

```python
import seaborn as sns
sns.set_style('darkgrid')
```

### Loading the data

```python
cars=sns.load_dataset('mpg')
cars.head()
```

### Creating a displot

```python
sns.displot(x='weight', data=cars);
```

```python
#we can also add kde line to it
sns.displot(x='weight', data=cars, kde=True);
```

```python
#Kind changing the type of plot
#We can change the histplot entirely in kdeplot
sns.displot(x='weight', data=cars, kind='kde');
```

```python
#Changing to ecdf plot
sns.displot(x='weight', data=cars, kind='ecdf');
```

```python
#Adding rugplot to the previously existing plot
sns.displot(x='weight', data=cars, kind='kde', rug=True);
#rugplot gives the 1D representation of kde
```

### Categorical Variables

```python
#Hue for third variable which is categorical
sns.displot(x='weight', data=cars, hue='origin');
#this kinda troubling since the colors are overlapping
```

```python
#Using kdeplot with hue
sns.displot(x='weight', data=cars, kind='kde', hue='origin');
#it makes it more easy to differentiate
```

### Bivariate (2D) displot

```python
sns.displot(x='weight',y='horsepower', data=cars);
#this is similiar to the heatmap
#here darker color represent larger values
```

```python
#instead of histplot we can used kde plot for bivariate representation
#it shows the relationship in terms of contour
sns.displot(x='weight',y='horsepower', data=cars, kind='kde');
```

```python
#even if we changed the type of plot we still have the access to the old plot
sns.displot(x='weight',y='horsepower', data=cars, kind='kde', rug=True);
```

### FacetGrid
Creating small multiples from displot

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin');
#we created seperate plots for each origin value so that all of them are easily visible
```

```python
#switching plots within the facetgrid
sns.displot(x='weight', data=cars, hue='origin', col='origin', kind='kde', rug=True);
```

```python
#Using both rows and columns of the facetgrid
#using cylinders for row
#filtering the cars data based on the number of cylinders
sns.displot(x='weight', data=cars[cars.cylinders.isin([4,6,8])], hue='origin', col='origin', row='cylinders');
```

### Styling

```python
#palette for changing the color of the plot
sns.displot(x='weight', data=cars, hue='origin', col='origin', palette='crest');
```

```python
#Height and Aspect for changing the size and the ratio respectively
sns.displot(x='weight', data=cars, hue='origin', col='origin', height=2, aspect=1.5);
```

our arguments of the plot depend on type of plot e.g. if you have histogram all of the options related to histogram will be readily avalaible for your displot.

```python
sns.displot(x='weight', data=cars, bins=20);
```

```python
#or if it is kde plot you can use options of the kdeplot
sns.displot(x='weight', data=cars, kind='kde', fill=True, lw=3);
```

```python
#rug_kws for making changes in the rugplot
sns.displot(x='weight', data=cars, kind='kde', rug=True, rug_kws={'height':0.05, 'color':'orange'});
```

```python
#FacetGrid Styling
#facet_kws
sns.displot(x='weight', data=cars, col='origin', facet_kws={'despine':False});
```

```python
#FacetGrid Object
g=sns.displot(x='weight', data=cars, col='origin');
```

```python
type(g)
#it is an facetgrid object
```
