# KDE - Kernel Density Estimation Plot
It is a continous analogous of histogram. It helps us estimate probablity density function from a finite data. It is created by putting kernel over each data point then add them density-wise (this addition looks like adding two waves in phase). There are various types of kernel such as triangular, guassiun (default) and sine.

```python
#importing libraries
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#setting style
sns.set_style('whitegrid')
```

```python
cars=sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

### Univariate KDE

```python
#KDE Plot
sns.kdeplot(cars.horsepower);
```

```python
#Adding Shade
sns.kdeplot(cars.horsepower, shade=True);
```

```python
#Adding Bandwidth : Increasing and decreasing bandwidth changes form of the KDE plot.
#As we increase the curve gets smoother.
sns.kdeplot(cars.horsepower, bw=0.1);
```

```python
#Insted of Probability Desnisty Function we get Cumulative Density Function from KDE plot
sns.kdeplot(cars.horsepower,cumulative=True);
```

### Bivariate KDE
We used for plotting joint probablity density function. It allows to make contours to represent various density levels. Closer the curve, higher density and vice-versa.

```python
sns.kdeplot(cars.horsepower,cars.mpg);
```

```python
#Adding more contours to the plot
sns.kdeplot(cars.horsepower,cars.mpg, n_levels=20);
```

```python
#Insted of lines we can use shades
sns.kdeplot(cars.horsepower,cars.mpg, shade=True);
```

```python
#Adding color bar to represent value associated with color
sns.kdeplot(cars.horsepower,cars.mpg, shade=True, cbar=True);
```

### Bivariate KDE by Category
In here we map the third variable with a hue semantic to show conditional distributions.

```python
#Relation between mpg, horsepower and number of cylinders.
sns.kdeplot(cars.horsepower,cars.mpg, shade=True, hue=cars.cylinders);
```
