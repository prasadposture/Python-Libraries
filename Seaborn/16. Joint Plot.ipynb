# Joint Plot
It helps to understand between two variables and the distribution of individual variable. In the middle it shows the relationship between x and y variable. And on the top and right hand side it shows the histogram of x and y data points repsectively. It gives the marginal distribution. It has six different relationships that could be shown in the middle the default one is scatterplot.

### Importing libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

### Loading the data

```python
geyser=sns.load_dataset('geyser')
geyser.head()
```

```python
geyser.shape
```

### Creating a joint plot

```python
sns.jointplot(data=geyser, x='waiting', y='duration');
```

# Plot kinds
1. scatter (default)
2. kde
3. reg
4. hist
5. hex
6. resid

```python
#Scatterplot
sns.jointplot(data=geyser, x='waiting', y='duration', kind='scatter');
```

```python
#KDE
sns.jointplot(data=geyser, x='waiting', y='duration', kind='kde');
```

```python
#Regplot
sns.jointplot(data=geyser, x='waiting', y='duration', kind='reg');
```

```python
#Histplot
sns.jointplot(data=geyser, x='waiting', y='duration', kind='hist');
```

### Working with JointGrid

```python
#Making additional changes in the plot
g=sns.jointplot(data=geyser, x='waiting', y='duration', kind='scatter');
```

```python
type(g)
```

```python
g.plot_joint(sns.kdeplot, color='gray', levels=5);
```

### Categorical Variables

```python
geyser.head()
```

```python
#Hue for categorical variables
sns.jointplot(data=geyser, x='waiting', y='duration', hue='kind');
```

```python
sns.jointplot(data=geyser, x='waiting', y='duration', hue='kind', kind='kde');
```

### Styling

```python
#Color
sns.jointplot(data=geyser, x='waiting', y='duration', color='green');
```

```python
#Palette
sns.jointplot(data=geyser, x='waiting', y='duration', hue='kind', palette='autumn');
```

```python
#Height for the size of the plot
#Space for spacing between marginal plot and plot in the middle 
#Ratio for adjusting the ratio of sizes of the inner and outer plot
sns.jointplot(data=geyser, x='waiting', y='duration', height=8, space=0.2, ratio=2);
```

### Working with Keywords

**Joint Keywords**   
`joint_kws` for making changes in the joint plot

```python
sns.jointplot(data=geyser, x='waiting', y='duration',joint_kws={'marker':5, 'color':'blue'});
#the keywords used are suitable for scatterplot only
```

```python
#similiarly if there is regplot in the middle are
#we can use keywords related to regplot
sns.jointplot(data=geyser, x='waiting', y='duration', kind='reg', joint_kws={'ci':None, 'order':2});
```

**Marginal Keywords**   
`marginal_kws` for making changes in the marginal plot

```python
sns.jointplot(data=geyser, x='waiting', y='duration', marginal_kws={'color':'xkcd:golden', 'ec':'black'});
#the keywords used here are suitable for histogram only since it's there as the marginal plot
```

```python
#similiary we can give different key word arguements for kde or any other plot
sns.jointplot(data=geyser, x='waiting', y='duration',kind='kde', marginal_kws={'lw':4, 'shade':True});
```
