# Color Palettes

```python
#import library
import seaborn as sns

#setting style
sns.set_style('whitegrid')

#load dataset
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day');
```

```python
#seaborn has use default color palette above
#we can check the default color palette as follows
sns.color_palette()
```

```python
#checking the other color palette
sns.color_palette('terrain')
```

```python
#changing the palette of the above plot
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', palette='terrain');
```

There are 170 palette names. You don't need to remember all of those, just give the palette argument whatever you the error message will give you the list of all valid color palettes. In the list of color palettes you will see `_r` many times that indicates the same color in reverse order.

```python
sns.color_palette('terrain_r')
```

### Extracting Colors as Variables

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, color='blue');
#gives matplotlib blue not seaborn blue
```

```python
blue, orange = sns.color_palette()[:2]
```

```python
blue
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, color=blue);
#gives the seaborn blue
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, color=orange);
#gives the seaborn orange
```

### Custom Palette

`.color_palette()`

```python
sns.color_palette('terrain', 12) #increasing the number of color palettes
```

`.light_palette()` and `.dark_palette()`

```python
sns.light_palette('blue')
```

```python
sns.dark_palette('blue')
```

`.blend_palette()`

```python
#blending two colors
sns.blend_palette(['blue','red'])
```

```python
#blending three colors and extending shades
sns.blend_palette(['blue','red','yellow'], 10)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', palette=sns.blend_palette(['red','blue'], 4));
```

### Highlighting Palette

```python
#creating palette dictionary
palette_dict = {x: 'gray' for x in tips.day.unique()}
palette_dict
```

```python
#giving highlighted color
palette_dict['Sat']='crimson'
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', palette=palette_dict);
```

```python
sns.displot(x='tip', data=tips , hue='day', col='day', palette=palette_dict);
```
