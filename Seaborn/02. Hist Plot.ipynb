# Hist Plot
This is used for plotting histogram in seaborn.

### Importing libraries and loading the data

```python
#importing libraries
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#setting style
sns.set_style('darkgrid')
#checking the version
sns.__version__
```

```python
penguins=sns.load_dataset('penguins')
penguins.head()
```

```python
#shape of the dataset
penguins.shape
```

```python
#dropping null values
penguins.dropna(inplace=True)
penguins.shape
```

### Creating histplot

```python
sns.histplot(penguins.bill_length_mm);
```

```python
#alternate way
sns.histplot(data=penguins,x='bill_length_mm');
```

```python
#plotting histograms horizontally
sns.histplot(data=penguins,y='bill_length_mm');
```

### Histplot over KDE

```python
sns.histplot(data=penguins, x='bill_length_mm',kde=True);
```

### Bins of histplot

```python
#Changing the number of bins
sns.histplot(data=penguins,x='bill_length_mm',bins=20);
```

```python
#Creating bins based on our choice
sns.histplot(data=penguins,x='bill_length_mm',bins=[30,40,50,60]);
```

```python
#Binwidth
sns.histplot(data=penguins,x='bill_length_mm',binwidth=10);
```

```python
#Binrange
sns.histplot(data=penguins,x='bill_length_mm', binwidth=10, binrange=(30,60));
```

### Histplot with Statistics

```python
#Histplot with count
sns.histplot(data=penguins,x='bill_length_mm',stat='count');
```

```python
#Hisplot with density (Density : Area of histogram sums to one)
sns.histplot(data=penguins,x='bill_length_mm',stat='density');
```

```python
#Histplot with Probability (Probability : Height of histogram bars to one )
sns.histplot(data=penguins,x='bill_length_mm',stat='probability');
```

```python
#Plotting histplot of cumulitive probability distribution
#Changing the style for probability histplot
sns.histplot(data=penguins,x='bill_length_mm',stat='probability',fill=False, element='step',cumulative=True);
```

### Dealing with Categorical Variables

```python
sns.histplot(data=penguins,x='bill_length_mm',hue='species');
```

```python
#changing the bars to steps using element arguement
#It makes easier to distinguish between the layers of the category
sns.histplot(data=penguins,x='bill_length_mm',hue='species',element='step');
```

```python
# We can also change the element to 'poly' to produce polygons
sns.histplot(data=penguins,x='bill_length_mm',hue='species',element='poly');
```

```python
#Using Multiple layer for different categories
sns.histplot(data=penguins,x='bill_length_mm',hue='species',multiple='layer');
```

```python
#We can also make use of multiple stack
sns.histplot(data=penguins,x='bill_length_mm',hue='species',multiple='stack');
```

```python
#We can also use multiple fill to fill the histogram with proportion of the presence of the variable value
sns.histplot(data=penguins,x='bill_length_mm',hue='species',multiple='fill');
```

### Bivariate Histplot

```python
#plotting a bivariate histplot
sns.histplot(data=penguins,x='bill_length_mm',y='bill_depth_mm');
#it kinda looks like heatmap
#hard to get any information from the colors though
```

```python
#getting the representation of the colors
sns.histplot(data=penguins,x='bill_length_mm',y='bill_depth_mm',cbar=True);
```

### Bivariate Histplot for Categorical Variable

```python
#Including categorical variable
sns.histplot(data=penguins,x='bill_length_mm',y='bill_depth_mm',hue='species');
#quite difficult to see the ovelap here
```

```python
#changing the variable
sns.histplot(data=penguins, x='bill_length_mm', y='species', hue='species');
#this looks neat
```

### Styling

```python
#Color : changing color of the bars
sns.histplot(data=penguins,x='bill_length_mm',color='green');
```

```python
#Fill=False : Gives only the outline of the bars
sns.histplot(data=penguins,x='bill_length_mm',fill=False);
```

```python
#Shrink : adds gap between the bars
#gives number of males and females for each species
sns.histplot(x='species',hue='sex',multiple='dodge', data=penguins, shrink=0.8);
```

```python
#Palette : adds multiple color suitable for the plot
sns.histplot(x='species',hue='sex',multiple='dodge',data=penguins, shrink=0.8, palette='bone');
```
