# Box Plot
It is used for representing the data in terms of percentile form. The box is made up of median, 25th and 75th percentile. The whisker are extended till $1.5 \times IQR$ (Inter Quartile Range) on both sides then they are pulled back to the last data point that lies within. All the other datapoints outside the box and not connected to whiskers represent outliers.

The boxplot gives stastical summary of the data in visual form also makes it easy to spot outliers in the data. We can compare to distribution of two attributes with the help of boxplot. Seaborn classfies boxplot as categorical distribution plot.

### Importing libraries

```python
#importing libraries
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#setting style
sns.set_style('darkgrid')
#checking the version
sns.__version__
```

### Reading the data

```python
#loading the data and dropping the null values
cars=sns.load_dataset('mpg').dropna()
cars.head()
```

```python
cars.shape
```

```python
cars.cylinders.value_counts()
```

Filter down to only cars with even number of cylinders.

```python
cars=cars[cars.cylinders.isin([4,6,8])]
cars.cylinders.value_counts()
```

### Creating a boxplot

```python
sns.boxplot(data=cars, x='mpg');
```

```python
cars['mpg'].describe()
```

You can clearly see the agreement between the values that we have got from the summary statistics and the values plotted in the boxplot.

### Multiple Boxplots

```python
# Boxplot based on category 
sns.boxplot(data=cars, x='origin',y='mpg');
```

```python
# Hue for different categories
sns.boxplot(data=cars, x='origin',y='mpg',hue='cylinders');
#looks like only usa origin cars has 8 cylinders
```

Suppose we need to classify cars as newer for `after 1976` and older for `before 1976` and see whose mpg is better.

```python
cars.model_year.describe()
```

```python
#this will add new column to our dataframe
cars['newer_model']=cars.model_year > 76
```

```python
#Creating boxplot for mpg for newer and older model
sns.boxplot(data=cars, x='origin', y='mpg', hue='newer_model');
```

Conclusion : Newer cars has good mpg

### Styling

```python
#Orientation
sns.boxplot(data=cars, y='origin', x='mpg');
```

```python
# Order : arranging the boxplots the way we want them
sns.boxplot(data=cars, x='origin', y='mpg',order=['japan','europe','usa']);
```

```python
#Hue Oreder : Changing the order of the hues
sns.boxplot(data=cars, x='origin', y='mpg', hue='newer_model', order=['japan','europe','usa'], hue_order=[True, False]);
```

```python
# Color : color of the boxplots
sns.boxplot(data=cars, x='origin', y='mpg',color='g');
```

```python
#Color variation with hue
sns.boxplot(data=cars, x='origin', y='mpg', color='g', hue='newer_model');
```

```python
#Width : changing width of the boxplot (default:0.8)
sns.boxplot(data=cars, x='origin', y='mpg', width=0.5);
```

```python
#Linewidth : changing linewidth of the boxplot (default:2.5)
sns.boxplot(data=cars, x='origin', y='mpg',linewidth=2.5);
```

```python
#Whis : for changing length of the whiskers (default:1.5)
sns.boxplot(data=cars, x='origin', y='mpg', whis=1);
#this changes the outliers
```

```python
#Fliersize : for changing filtersize (sizes of the dianmond denoting the outliers) (default:5)
sns.boxplot(data=cars, x='origin', y='mpg', fliersize=2);
```

```python
#Showcaps : removes the caps from the whiskers
sns.boxplot(data=cars, x='origin', y='mpg', showcaps=False);
```
