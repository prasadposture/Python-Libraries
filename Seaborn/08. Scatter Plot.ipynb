# Scatter Plot
It is plot of two dimensional points on xy plane. It is a relational plot. It helps us find the correlation between the data points.

### Importing libraries

```python
#importing libraries
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#setting style
sns.set_style('darkgrid')
#checking the version
sns.__version__
```

### Reading the data

```python
#loading the data and dropping the null valuesd
diamonds=sns.load_dataset('diamonds')
diamonds.head()
```

```python
diamonds.shape
```

Filtering the data

```python
diamonds=diamonds[
    diamonds.cut.isin(['Premium','Good']) &
    diamonds.color.isin(['D','F','J'])
].sample(n=100, random_state=22)
diamonds.shape
```

```python
diamonds['cut'].unique
```

### Creating a scatter plot

```python
#plot of carat vs price
sns.scatterplot(data=diamonds, x='carat', y='price');
```

```python
#plot of x v
sns.scatterplot(data=diamonds, x='x', y='y');
```

### Plotting with third variable
Hue  
Size  
Style  

```python
#Hue
sns.scatterplot(data=diamonds, x='carat', y='price', hue='cut');
```

```python
#Palette : changing the color of the hue
sns.scatterplot(data=diamonds, x='carat', y='price', hue='cut', palette=['red','orange','green','blue','black']);
```

```python
#Hue for continuous variable variable
sns.scatterplot(data=diamonds, x='carat', y='price', hue='depth');
```

```python
#Style : for categorical variable
sns.scatterplot(data=diamonds, x='carat', y='price',s=100,  style='cut');
```

```python
#Size: for plotting categorical variable
sns.scatterplot(data=diamonds, x='carat', y='price', size='cut');
```

```python
#specified sizes
sns.scatterplot(data=diamonds, x='carat', y='price', size='cut', sizes=[10,150,10,50,10]);
```

```python
#Using size for continuous variable
sns.scatterplot(data=diamonds, x='carat', y='price', size='depth');
```

```python
#Style+Size+Hue for the data point
sns.scatterplot(data=diamonds, x='carat', y='price', hue='cut', style='cut', s=100);
```

```python
#Color and Cut of the diamond through style and hue
sns.scatterplot(data=diamonds, x='carat', y='price', hue='cut', style='color', s=100);
```

### Styling

```python
#Transperency of the dots by alpha
sns.scatterplot(data=diamonds, x='carat', y='price', alpha=0.5, s=100);
```

```python
#Marker
sns.scatterplot(data=diamonds, x='carat', y='price', alpha=0.5, s=100, marker='*');
```

```python
#Changing color of the outline of the data points
sns.scatterplot(data=diamonds, x='carat', y='price', alpha=0.5, s=100, marker='*',edgecolor='black');
```
