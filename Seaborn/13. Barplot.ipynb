# Bar Plot
It's a categorical plot used for plotting mean of the values related to the particular category on y-axis.

### Import libraries

```python
#importing libraries
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#setting style
sns.set_style('darkgrid')
```

### Reading the data

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.shape
```

### Creating a bar plot

```python
sns.barplot(data=penguins, x='species', y='body_mass_g');
```

we can get the means of these categories by using `.groupby()` and verify the answer by ourselves.

```python
penguins.groupby('species').body_mass_g.mean()
```

```python
#changing the orientation
sns.barplot(data=penguins, y='species', x='body_mass_g');
```

we can also add confidence interval to our barplot which is done using bootstrap method.

```python
#adding confidence interval (default = 95)
sns.barplot(data=penguins, x='species', y='body_mass_g', ci=68);
```

```python
#We can turn this confidence interval off by using None
sns.barplot(data=penguins, x='species', y='body_mass_g', ci=None);
```

```python
#Estimator
#Instead of mean we can use the standard deviation to plot the bars
sns.barplot(data=penguins, x='species', y='body_mass_g', estimator=np.std);
```

```python
#We can verify this by using .describe()
penguins.groupby('species').body_mass_g.describe()
```

```python
#using max as the estimator
sns.barplot(data=penguins, x='species', y='body_mass_g', estimator=np.max);
```

### Plotting with two categorical variables

```python
sns.barplot(data=penguins, x='species', y='body_mass_g', hue='sex');
```

```python
#Horizontal barplot
sns.barplot(data=penguins, y='species', x='body_mass_g', hue='sex');
```

```python
#Changing the order of the bars
sns.barplot(data=penguins, x='species', y='body_mass_g', order=['Gentoo', 'Chinstrap', 'Adelie']);
```

```python
#changing the order of the hue
sns.barplot(data=penguins, x='species', y='body_mass_g', hue='sex', hue_order=['Female', 'Male']);
```

### Styling

```python
#Color of the bars
sns.barplot(data=penguins, x='species', y='body_mass_g', color='c');
```

```python
#Changing the color and width of the error
sns.barplot(data=penguins, x='species', y='body_mass_g', color='c', errcolor='g', errwidth=8);
```

```python
#Palette
sns.barplot(data=penguins, x='species', y='body_mass_g', palette='hot');
```

```python
#changing edgecolor and linewidth
sns.barplot(data=penguins, x='species', y='body_mass_g', color='aliceblue', edgecolor='mediumblue', lw=4);
```
