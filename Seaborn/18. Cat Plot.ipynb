# Cat Plot
It stands for categorical plot. With this we can all the categorical plots offered by seaborn. It also provides access to FacetGrid methods.

### Import libraries

```python
import seaborn as sns
sns.set_style('darkgrid')
```

### Loading and Filtering

```python
#loading the data
cars=sns.load_dataset('mpg').dropna()

#filtering for 2,6,8 cylinders
cars=cars[cars.cylinders.isin([4,6,8])]

#add old/new column
cars['type']=['old' if x<76 else 'new' for x in cars.model_year]
```

```python
cars.head()
```

```python
cars.shape
```

### Creating a catplot

```python
sns.catplot(data=cars, x='horsepower');
#the default catplot is actually a strip plot
```

```python
#changing the orientation
sns.catplot(data=cars, y='horsepower');
```

```python
#changing the kind to boxplot
sns.catplot(data=cars, x='horsepower', kind='box');
```

```python
#changing the kind to violin plot
sns.catplot(data=cars, x='horsepower', kind='violin');
```

### Categorical variables

```python
#categorical variable
sns.catplot(data=cars, x='origin', y='horsepower', kind='box');
```

```python
#Hue for third categorical variable
sns.catplot(data=cars, x='origin', y='horsepower',hue='cylinders', kind='box');
```

```python
#changing to violin plot
sns.catplot(data=cars, x='origin', y='horsepower',hue='cylinders', kind='violin');
```

### Small Multiples (Facet Grid)

```python
#columns
sns.catplot(data=cars, x='cylinders', y='horsepower',col='origin', kind='box');
```

```python
#changing the above boxplots to striplots
sns.catplot(data=cars, x='cylinders', y='horsepower',col='origin', kind='strip');
```

```python
#columns and rows
sns.catplot(data=cars, x='cylinders', y='horsepower',col='origin',row='type', kind='strip');
```

```python
#changing the order of the facet grid rows
sns.catplot(data=cars, x='cylinders', y='horsepower',col='origin',row='type', kind='strip', row_order=['new','old']);
```

### Styling
In general, additional keyword arguments will be passed to the kind of plot selected.

```python
#Boxplot stylings
#palette
#whisker length
sns.catplot(data=cars, x='cylinders', y='horsepower',col='origin', kind='box', palette='winter', whis=1);
```

```python
#Stripplot stylings
#size of the dots
#Transperency by alpha
#adding line around the data points
sns.catplot(data=cars, x='cylinders', y='horsepower',col='origin', kind='strip', s=12, alpha=0.5, linewidth=1);
```

### Facet Grid Styling

```python
g=sns.catplot(data=cars, x='cylinders', y='horsepower',col='origin', kind='box');
```

```python
type(g)
```

```python
#changing the font scale
sns.set(style='darkgrid', font_scale=1.5)

#facetgrif
g=sns.catplot(data=cars, x='cylinders', y='horsepower',col='origin', kind='box')
g.set_titles(col_template='{col_name} cars');
```
