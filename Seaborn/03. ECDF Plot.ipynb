# ECDF Plot
It is stands for Emperical Cumulative Distribution Function. It is plot values arranged in ascending order and their proportion (based on the number of time an observation has occurred and total number of observations) in the data. It ranges form minimum to the maximum of the data. <br>

**Advantages:**
1. No binning or smoothing
2. Easy to compare category distributions

**Disadvantages:**
1. Difficult to detect central tendencies
2. Not suitable for bimodial distributions

### Importing libraries

```python
#importing libraries
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#setting style
sns.set_style('darkgrid')
#checking the version
sns.__version__
```

### Reading the data

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

### Creating ecdf plot

```python
sns.ecdfplot(data=tips,x='tip');
```

```python
tips.tip.min()
```

```python
tips.tip.max()
```

```python
#since there is a long straight line at 2, that implies a significant proportion points are equal to 2
tips.tip.value_counts().head()
```

```python
#adding a vertical line for showing clear distnictions
sns.ecdfplot(data=tips,x='tip')
plt.axvline(4,c='black');
#Observation : nearly 80% of the tip value is less than 4 dollors and 20% is greater than 4 dollors
```

```python
#Changing the orientation of the plot
sns.ecdfplot(data=tips,y='tip')
plt.axvline(0.5,c='black'); #for 50th percentile of the data
```

```python
#checking the 50th percentile aka median
tips.tip.median()
#agreement
```

### Additional Options

```python
#hue for categoric variables
#comparison of tips based on  time : dinner / lunch
sns.ecdfplot(data=tips,x='tip',hue='time');
```

```python
#comparison of tips based on  day of the week
sns.ecdfplot(data=tips,x='tip',hue='day');
```

```python
#count for actual values instead of thier proportion in the data
sns.ecdfplot(data=tips,x='tip',stat='count');
```

```python
#using hue and stat
sns.ecdfplot(data=tips,x='tip',stat='count',hue='day');
```

```python
# Ranking with complementary and count
sns.ecdfplot(data=tips,x='tip',stat='count',complementary=True);
#10 dollar tip has first rank
#2 dollar tip has 200th rank
```

```python
#Weights : We use weights for weighing the proportion.
#Here 50% of the tip money made, came from tips less than about $3.50 
sns.ecdfplot(data=tips,x='tip',weights='tip')
plt.axhline(0.5,c='black');
```

### Styling

```python
#Palette : giving various shades of colors instead of various colors
sns.ecdfplot(data=tips,x='tip',hue='day',palette='summer');
```

```python
#Changing the linewidth
sns.ecdfplot(data=tips,x='tip',hue='day',lw=3);
```

```python
# Editing the legend
# Adding name list
# Adding title to the legend
# Changing the location
p=sns.ecdfplot(data=tips,x='tip',hue='day')
p.legend(['Thursday','Friday','Saturday','Sunday'], title='Day of week',loc=4);
```
