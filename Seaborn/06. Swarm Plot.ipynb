# Swarm Plot
The data is first sorted in ascending order then circular dots are plotted on the axis with the frequency of the data points. This also sometimes called Bee Swarm Plot. It considered as the categorical scatter plot.

### Importing libraries

```python
#importing libraries
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#setting style
sns.set_style('darkgrid')
#checking the version
sns.__version__
```

### Reading data

```python
#loading the data and dropping the null values
cars=sns.load_dataset('mpg').dropna()
cars.head()
```

```python
cars.shape
```

```python
#Cleaning the data
cars.cylinders.value_counts()
```

Filter down to only cars with even number of cylinders.

```python
cars=cars[cars.cylinders.isin([4,6,8])]
cars.cylinders.value_counts()
```

### Creating swarmplot

```python
sns.swarmplot(data=cars,x='horsepower');
```

```python
#spltting data for swarmplot based on the category
sns.swarmplot(data=cars,x='origin',y='horsepower');
```

```python
#plotting the third categorical variable using hue
sns.swarmplot(data=cars,x='origin',y='horsepower',hue='cylinders');
```

```python
#Seperating the groups classified on the basis of hue
sns.swarmplot(data=cars,x='origin',y='horsepower',hue='cylinders', dodge=True);
```

### Swarmplot overlay

```python
usa=cars[cars.origin=='usa']
```

#### Swarmplot with Boxplot

```python
sns.boxplot(data=cars,x='cylinders',y='horsepower')
sns.swarmplot(data=cars,x='cylinders',y='horsepower');
#but this representation isn't quite pleasing to the eyes
```

```python
#Changing the color of the dots and making them transperent
#Using color and alpha
sns.boxplot(data=cars,x='cylinders',y='horsepower')
sns.swarmplot(data=cars,x='cylinders',y='horsepower',color='black',alpha=0.5);
```

```python
#Changing the size of the whsikers
sns.boxplot(data=cars,x='cylinders',y='horsepower',whis=np.inf)
sns.swarmplot(data=cars,x='cylinders',y='horsepower',color='black',alpha=0.5);
```

#### Swarmplot with Violin plot

```python
sns.violinplot(data=cars,x='cylinders',y='horsepower')
sns.swarmplot(data=cars,x='cylinders',y='horsepower');
#Messy representation again
```

```python
#Making the scale of all violins same and changing the color and size of the dots
#Extending width takes all the points within the violin
#Minimizing the size of the dots helps accumulate them altogether
sns.violinplot(data=cars,x='cylinders',y='horsepower',scale='width')
sns.swarmplot(data=cars,x='cylinders',y='horsepower',color='white',size=3);
```

```python
#Turning off the boxportion of the violin plot with inner
#Changing the color of the violinplot with dark color so that white dots are easily visible
sns.violinplot(data=cars,x='cylinders',y='horsepower',scale='width',inner=None, color='green')
sns.swarmplot(data=cars,x='cylinders',y='horsepower',color='white',size=3);
```

### Styling

```python
#Line width and Edgecolor
#Adds outline to the each dot
#Edgecolor changes the color of the outline
sns.swarmplot(data=cars,x='horsepower',linewidth=1, edgecolor='red');
```

### Matplotlib scatter properties

```python
#Marker
plt.figure(figsize=(8,4))
sns.swarmplot(data=cars,x='horsepower',marker='*');
```
