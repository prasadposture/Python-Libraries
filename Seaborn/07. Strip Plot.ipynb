# Strip Plot
For creating a strip plot we take a blank line and each point is plotted on the line from the data set. In this plot the repeated data point won't be plotted above the previously existing point like we had it in swarmplot, but we can use `jitter` to seperate this point little above and below. It is a categorical scatterplot. Since points plotted in strip plot are transperent you can clearly distinguish between them based on the intesity. It can represent a lot more data than the swarmplot.

### Importing libraries

```python
#importing libraries
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#setting style
sns.set_style('darkgrid')
#checking the version
sns.__version__
```

### Reading the data

```python
#loading the data and dropping the null values
cars=sns.load_dataset('mpg').dropna()
cars.head()
```

```python
cars.shape
```

```python
#Cleaning the data
cars.cylinders.value_counts()
```

Filter down to only cars with even number of cylinders.

```python
cars=cars[cars.cylinders.isin([4,6,8])]
cars.cylinders.value_counts()
```

### Creating a strip plot

```python
sns.stripplot(cars.weight);
```

```python
sns.stripplot(data=cars, x='weight',y='origin');
```

```python
#Using hue for third parameter
sns.stripplot(data=cars, x='weight',y='origin',hue='cylinders');
```

```python
#Seperating the data points based on hue
sns.stripplot(data=cars, x='weight',y='origin',hue='cylinders',dodge=True);
```

```python
#Changing the orientation
sns.stripplot(data=cars, y='weight',x='origin');
```

### Styling

```python
#Jitter : It's on by default
#Let's see what happens if we turned it off
sns.stripplot(data=cars, x='weight',y='origin',jitter=False);
#it's difficult for us to see the data points
```

```python
#Changing the jitter partially
sns.stripplot(data=cars, x='weight',y='origin',jitter=0.05);
```

```python
#Changing jitter size using widget
from ipywidgets import interactive
def jitter_widget(jitter=0.1):
    sns.stripplot(cars.weight, cars.origin, jitter=jitter)
```

```python
interactive(jitter_widget, cars.origin, jitter=(0,0.65,0.05))
```

```python
#Checking transperency of the dots
sns.stripplot(cars.weight, cars.origin, alpha=0.5);
```

```python
#Changing the size of the dots
sns.stripplot(cars.weight, cars.origin, alpha=0.3, size=8);
```

```python
#Linewidth : adding outlie to the data points
sns.stripplot(cars.weight, cars.origin, linewidth=1);
```

```python
#Edgecolor : changing color of the outline
sns.stripplot(cars.weight, cars.origin, linewidth=1, edgecolor='black');
```

```python
#Marker
sns.stripplot(cars.weight, cars.origin, marker='*', size=10, alpha=0.7, jitter=0.25);
```
