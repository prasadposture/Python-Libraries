# Pair Plot
It used for representing the correlation between the variables in various forms. It plots the scatterplot or regplot on the off diagonal and histogram or KDE on diagonals. Default setting is, scatterplot on the off diagonal and histogram on diagonals.

### Importing libraries

```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
sns.set_style('darkgrid')
```

### Loading the data

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
sns.pairplot(tips);
```

```python
#KDE plot on diagonal
sns.pairplot(tips, diag_kind='kde');
```

```python
#Regplot on off diagonals
sns.pairplot(tips, kind='reg');
```

### Pairplots returns PairGrid
We can store this into an variables and make changes.

```python
g=sns.pairplot(tips);
```

```python
type(g)
```

```python
#saving the pairgrid
g=sns.pairplot(tips)

#adding to the pairgrid
g.map_upper(sns.kdeplot, n_levels=6, color='xkcd:wine red');
#adds kdeplot to the upper triangle
```

### Dealing with Categorical Variables

```python
#data cleaning
tips['weekend']=tips.day.isin(['Sat', 'Sun'])
tips.head()
```

```python
#Hue for categorical variables
sns.pairplot(tips, hue='weekend');
#blue for none weekend
#orange for weekend
```

```python
#Mentioning the variables we want to see the pairplot of.
sns.pairplot(tips, hue='weekend', vars=['total_bill', 'tip']);
```

```python
#We can also speicify the variables for the axis
sns.pairplot(tips, hue='weekend', y_vars=['tip'], x_vars=['total_bill', 'size'], kind='reg', height=4);
```

### Advanced Styling

```python
#we will delete the weekend column that we create
del tips['weekend']
```

**Height and Aspect**

```python
#We will add widgets so that the plots become interactive
from ipywidgets import interactive
def size_widget(height=2.5, aspect=1):
    sns.pairplot(tips, height=height, aspect=aspect)
```

```python
interactive(size_widget, height=(1, 3.5, 0.5), aspect=(0.5, 2, 0.25))
#height for changing the size of the plot
#aspect for changing the aspect ratio
```

**Palette**

```python
sns.pairplot(tips, hue='day', palette='plasma');
```

### Different Key Words

```python
#Diagonal Keywords (diag_kws) for changes in diagonal plots
sns.pairplot(tips, diag_kws={'color':'gray', 'alpha':0.6});
```

```python
#Plot Keywords for making changes in the non diagonal plots
sns.pairplot(tips, kind='reg', plot_kws={'ci':None, 'color':'xkcd:salmon', 'scatter_kws':{'color':'gray', 'alpha':0.5}})
```
