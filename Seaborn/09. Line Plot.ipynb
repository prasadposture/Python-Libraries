# Line Plot
It is use for showing the relation between two continuous variables. If data points are discrete they are connected with lines. It makes it easy to detect patterns and trends. With the line seaborn can also give you the confidence interval which is useful in understanding the spread of the values across the mean and making predictions. 

### Importing libraries

```python
#importing libraries
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#setting style
sns.set_style('darkgrid')
#checking the version
sns.__version__
```

https://archive.ics.uci.edu/ml/machine-learning-databases/00482/dataset.zip

```python
df = pd.read_csv('dataset.csv', parse_dates=[3])
df.head()
```

```python
#Renaming the columns
df.rename(columns={'SystemCodeNumber': 'Location', 'LastUpdated': 'Timestamp'}, inplace=True)
#Creating new columns
df['Day'] = df.Timestamp.dt.date
df['Month'] = df.Timestamp.dt.month
df['Hour'] = df.Timestamp.dt.hour
```

```python
#Creating new data frame for our use
park = df[df.Location.isin(['Broad Street', 'NIA South'])]
park.head()
```

```python
#Creating regular intervals
months = [pd.datetime(2016, 10, 1), 
          pd.datetime(2016, 11, 1), 
          pd.datetime(2016, 12, 1)]
plt.rc('date.autoformatter', day='%b %Y');
```

### Creating a lineplot

```python
sns.lineplot(data=park,x='Day',y='Occupancy');
#the months seem to be repeating
```

```python
#Adding xticks
sns.lineplot(data=park,x='Day',y='Occupancy')
plt.xticks(months);
```

```python
#Hour vs Occupancy
sns.lineplot(data=park,x='Hour',y='Occupancy');
```

```python
#Changing the amount of bootstrapping
#n_boot=1000 (default)
sns.lineplot(data=park,x='Hour',y='Occupancy', n_boot=10);
```

```python
#Changing the size of the confidence interval
#68%
sns.lineplot(data=park,x='Hour',y='Occupancy',ci=68);
```

```python
#Turning off the confidence interval
sns.lineplot(data=park,x='Hour',y='Occupancy',ci=None);
```

```python
#Here the line plot is by default plotting against mean instead we can use other stastical measure such as sum
sns.lineplot(data=park,x='Hour',y='Occupancy',estimator='sum');
```

```python
#Seeing standard deviation as an estimator instead of mean
sns.lineplot(data=park,x='Hour',y='Occupancy',estimator='std');
```

```python
#Splitting the plot with category
sns.lineplot(data=park,x='Day', y='Occupancy', hue='Location')
plt.xticks(months);
```

```python
#Giving colors of our own choice 
sns.lineplot(data=park,x='Day', y='Occupancy', hue='Location',palette=['red','green'])
plt.xticks(months);
```

```python
#Style : changing the style of the line for different categories
sns.lineplot(data=park,x='Day', y='Occupancy', style='Location')
plt.xticks(months);
```

```python
#Style and Hue together
sns.lineplot(data=park,x='Day', y='Occupancy', style='Location', hue='Location')
plt.xticks(months);
```

```python
#Size for a category
sns.lineplot(data=park,x='Hour', y='Occupancy', size='Location');
```

```python
#We can plot multiple lines with varying size
sns.lineplot(data=park,x='Hour', y='Occupancy', size='Month', ci=None) #Turning CI off so that all lines will clearly visibl
```

```python
#Changing the linewidth
sns.lineplot(data=park,x='Hour',y='Occupancy',ci=None, lw=4);
```

```python
#changing the color of the line
sns.lineplot(data=park,x='Hour',y='Occupancy',ci=None, color='black');
```

```python
#changing the transperancy of the line
sns.lineplot(data=park,x='Hour',y='Occupancy',ci=None, alpha=0.5);
```
