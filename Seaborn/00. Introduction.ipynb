### Comparison of Matplotlib and Seaborn

#### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=[0,15,10,25]
```

```python
plt.plot(x,y)
```

#### Seaborn
Seaborn updates `matplolib`'s rc parameters to improve aesthetics.

```python
import seaborn as sns
```

```python
sns.set()
```

```python
plt.plot(x,y);
```

```python
#this sets the style of plot similiar to the matplotlib
sns.set_style('white')
```

```python
plt.plot(x,y)
```

#### Versions

```python
#Version of matplotlib
import matplotlib
matplotlib.__version__
```

```python
sns.__version__
```

Seaborn updates maplotlib's rc:run commands parameters. These are default parameters which are intialized when we plot new figure such as background color, grid, fontsize etc. Seaborn is a rapper for matplotlib i.e. seaborn is built over matplotlib code. It offers better functionality, syntax and defaults.

### Seaborn with Pandas
Let seaborn group, aggregate and plot your `pandas` dataframes.

```python
cars = sns.load_dataset('mpg')
```

```python
type(cars)
```

```python
cars.dropna(inplace=True)
cars.shape
```

```python
cars.head()
```

```python
#We will see the explaination of this code later
sns.relplot(x='model_year',y='mpg',col='origin',hue='cylinders',data=cars);
```
