# Joint Grid
It is the backbone of the JointPlot. It is similiar to the FacetGrid and PairGrid. It consists of two plots joint distribution in the middle and marginal distribution at the edges.

```python
# import libraries
import seaborn as sns

#setting style
sns.set_style('darkgrid')

#load dataset
cars=sns.load_dataset('mpg')
cars.head()
```

### Creating a JointGrid

```python
g=sns.JointGrid();
```

```python
type(g)
```

```python
g=sns.JointGrid(data=cars,x='weight', y='mpg');
#nothing has happen yet
#cause we havent plotted any function on it
```

`.plot()`

```python
g=sns.JointGrid(data=cars,x='weight', y='mpg')
g.plot(sns.scatterplot, sns.histplot);
```

```python
#regplot and kdeplot
g=sns.JointGrid(data=cars,x='weight', y='mpg')
g.plot(sns.regplot, sns.kdeplot);
```

### Joint and Marginal Plots

`.plot_joint()`, `.plot_marginals()`

```python
g=sns.JointGrid(data=cars,x='weight', y='mpg')
g.plot_joint(sns.regplot)
g.plot_marginals(sns.kdeplot, fill=True);
#it is useful for making individual changes in the plots
```

### JointGrid Axes

```python
type(g.ax_marg_x)
```

```python
g=sns.JointGrid()
x,y=cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax=g.ax_joint)
sns.kdeplot(x=x, ax=g.ax_marg_x)
sns.histplot(y=y, ax=g.ax_marg_y);
```

### Hue for categorical data

```python
g=sns.JointGrid(data=cars,x='weight', y='mpg', hue='origin')
g.plot(sns.scatterplot, sns.kdeplot);
#gives hue for all three plots
```

```python
#hue for individual plots
g=sns.JointGrid()
x,y=cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax=g.ax_joint)
sns.kdeplot(x=x, ax=g.ax_marg_x, hue=cars.origin)
sns.histplot(y=y, ax=g.ax_marg_y);
```

### Styling

```python
#General styling
g=sns.JointGrid(data=cars, x='weight', y='mpg')
g.plot(sns.scatterplot, sns.histplot, alpha=0.8, edgecolor='black')
```

```python
#Plot specific keywords
g=sns.JointGrid(data=cars,x='weight', y='mpg')
g.plot_joint(sns.regplot, line_kws={'color':'black'})
g.plot_marginals(sns.kdeplot, color='purple', fill=True);
```

```python
#JointGrid Properties
g=sns.JointGrid(data=cars,x='weight', y='mpg')
g.plot(sns.scatterplot, sns.kdeplot)
#setting axis
g.set_axis_labels(xlabel='Weight (lb)', ylabel=' Miles per Gallon', fontsize=14)
#saving figure
g.savefig('JointGrid.png');
```
