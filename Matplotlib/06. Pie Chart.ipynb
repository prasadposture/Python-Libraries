# Pie Chart
A pie-chart is a representation of values as slices of a circle with different colors. It is created using `pie()` function from matplolib

```python
#Importing the library
import matplotlib.pyplot as plt
```

### Creating a pie chart

```python
#values to be presented
Task=[300,500,700]
#plotting with pie()
plt.pie(Task)
plt.show()
#but this doesn't give any information about what the values talk about
```

### Adding labels
This helps us understanding which portion is assigned to what.

```python
Task=[300,500,700] #values
my_label=['Task Pending','Task Oungoing','Task Completed'] #labels to the pie chart
plt.pie(Task,labels=my_label) #pie()
plt.show()
```

### Adding percentage
This helps us understand the pie chart better since we can easily compare when the slices are indinstiguishable.

```python
Task=[300,500,700]
my_label=['Task Pending','Task Oungoing','Task Completed']
plt.pie(Task,labels=my_label,autopct="%1.1f%%") #autopct is used for adding percentage 
plt.show()
```

### Adding title and axis

```python
Task=[300,500,700]
my_label=['Task Pending','Task Oungoing','Task Completed']
plt.pie(Task,labels=my_label,autopct="%1.2f%%")#two digits after decimal in calculated percentage
plt.title('My Task',fontsize=14)
plt.axis('equal')
plt.show()
```

```python
#values and labels
exp_label=['Home Rent','Food','Phone Internet Bill','Car','Other Utilities']
exp_vals=[1400,600,300,500,250]
#pie chart
plt.pie(exp_vals,labels=exp_label,autopct="%1.1f%%")
#title and axis
plt.title('My Expenses',fontsize=14)
plt.axis('equal')
plt.show()
```

### Changing the start angle
The plotting on the pie start from the positive x axis if we kept its center at the origin. By adding start angle we can change the position where the plotting begins.

```python
#values and labels
exp_label=['Home Rent','Food','Phone Internet Bill','Car','Other Utilities']
exp_vals=[1400,600,300,500,250]
#pie char
plt.pie(exp_vals,labels=exp_label,autopct="%1.1f%%",startangle=60)
#title and axis
plt.title('My Expenses',fontsize=14)
plt.axis('equal')
plt.show()
```

### Adding colors of our own choice

```python
#labels, titles and colors
exp_label=['Home Rent','Food','Phone Internet Bill','Car','Other Utilities']
exp_vals=[1400,600,300,500,250]
my_colors=['lightblue','darkgreen','lightsteelblue','silver','orange']
#plotting the pie chart
plt.pie(exp_vals,labels=exp_label,autopct="%1.1f%%",startangle=30,colors=my_colors)
#title and axis
plt.title('My Expenses',fontsize=14)
plt.axis('equal')
plt.show()
```

### Adding shadows
Adding shadows makes the plot look 3D.

```python
#labels, titles and colors
exp_label=['Home Rent','Food','Phone Internet Bill','Car','Other Utilities']
exp_vals=[1400,600,300,500,250]
my_colors=['lightblue','darkgreen','lightsteelblue','silver','orange']
#plotting the pie chart
plt.pie(exp_vals,labels=exp_label,autopct="%1.1f%%",startangle=30,colors=my_colors,shadow=True)
#title and axis
plt.title('My Expenses',fontsize=14)
plt.axis('equal')
plt.show()
```

### Seperating the slices with explode

```python
#labels, titles, colors and explode
exp_label=['Home Rent','Food','Phone Internet Bill','Car','Other Utilities']
exp_vals=[1400,600,300,500,250]
my_colors=['lightblue','darkgreen','lightsteelblue','silver','orange']
my_explode=(0,0,0,0.2,0) #cutting slice of forth value
#plotting the pie chart
plt.pie(exp_vals,labels=exp_label,autopct="%1.1f%%",startangle=30,colors=my_colors,shadow=True,explode=my_explode)
#title and axis
plt.title('My Expenses',fontsize=14)#giving title
plt.axis('equal')#adding axis
plt.show()
```

### Pie Chart with Pandas Data Frame

```python
from pandas import DataFrame
```

```python
Data= {
    'Task':[300,500,700]
}
df=DataFrame(Data, columns=['Task'])
print(df)
```

```python
my_label=['Task Pending','Task Oungoing','Task Completed']
#plotting with df['name of the attribute']
plt.pie(df['Task'],labels=my_label)
plt.title('My Task',fontsize=14)
plt.axis('equal')
plt.show()
#we can add all the details to this pie chart as we did before
```

```python
#Example

labels=["Python","C++","Ruby","Java"]
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral']
explode=(0.4,0,0,0)

plt.pie(sizes,
        labels=labels,
        autopct="%2.4f%%",
        startangle=30,
        colors=colors,
        shadow=True,
        explode=explode)

plt.title('My Task',fontsize=14)
plt.axis('equal')
plt.show()
```
