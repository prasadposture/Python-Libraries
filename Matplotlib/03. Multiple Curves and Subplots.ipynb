# Multiple Curves and Subplots

#### Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Plotting multiple curves and adding legend

```python
#Example
x=np.array([1,2,3,4])
y=2*x^2 #function 1
z=3*x^3 #function 2

#first plot with x and y
plt.plot(x,y,label='y=2*x^2')

#second plot with x and z
plt.plot(x,z,label='z=3*x^3')

#labelling
plt.xlabel("X axis")
plt.ylabel("Y axis")

#title
plt.title("Multiple Plots")

#adding a legend
plt.legend()
#acts as an indicator to show which equation is of which color
#a box in the corner to the left
```

#### Creating Subplots
Subplots make it easy to compare between various plots since it keeps them side by side.

```python
#plot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)#position of first subplot
plt.plot(x,y)

#plot 2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2) #position of second subplot
plt.plot(x,y)
```

```python
#another example of subplots

x=np.arange(0,10)
y=x**2 #function

plt.subplot(2,2,1) #position of subplot
plt.plot(x,y,'r--') #red dashed lines

plt.subplot(2,2,2) #position of subplot
plt.plot(x,y,'g*--') #green dashed and starred linne

plt.subplot(2,2,3) #position of subplot
plt.plot(x,y,'bo') #blue dots

plt.subplot(2,2,4) #position of subplot
plt.plot(x,y,'go') #green dots
```

#### Adding Title (for each subplot) and Supetitle (common for all of them)

```python
#plot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Sale')

#plot 2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.title('Income')

#Suptitle
plt.suptitle('My shop')#supertitle of the titles
```

#### Plotting Sin and Cos Waves

```python
#value of pi
np.pi
```

Sine Wave

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sin Wave')
plt.plot(x,y)
```

Subplots of Sin and Cos Waves

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('Sin Wave')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cos Wave')

plt.suptitle('Plot of the waves')
```

Observing the phase difference by plotting both curves on the same graph

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

We can also change the phase of sin and obtained beautiful pattern.

```python
plt.plot(x,np.sin(x-0),color="g",label='sin(x-0)')
plt.plot(x,np.sin(x-1),color="b",label='sin(x-1)')
plt.plot(x,np.sin(x-2),color="r",label='sin(x-2)')
plt.plot(x,np.sin(x-3),color="black",label='sin(x-3)')
plt.plot(x,np.sin(x-4),color="grey",label='sin(x-4)')
plt.plot(x,np.sin(x-5),color="yellow",label='sin(x-5)')
plt.legend()
```
