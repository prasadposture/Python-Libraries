```python
import matplotlib.pyplot as plt
```

```python
import numpy as np #import these two libraries in every line of the code
```

```python
#creating scatter plot
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)
plt.show()
```

```python
#compare plots
#day ONE, the age and speed of 13 cars
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)

#day TWO, the age and speed of 13 cars
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y)
```

```python
#change the colors
#day ONE, the age and speed of 13 cars
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y,color='hotpink')

#day TWO, the age and speed of 13 cars
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y,color='#88c999')
```

```python
#Labelling the axis and giving a title to the graph
x=np.arange(0,10)
y=np.arange(11,21)
plt.scatter(x,y,c='green')#one can change it red as well
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph in 2D")
plt.savefig('Test1.png')
```

```python
#giving each pixel a different color
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors=np.array(['red','green','blue','yellow','orange','cyan','beige','skyblue','pink','brown','black','gray','purple'])
plt.scatter(x,y,c=colors)
```

```python
#Size of the pixels
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,77])
plt.scatter(x,y,s=sizes)
```

```python
#Aplha
#adjusting the transperancy of the colors
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,77])
plt.scatter(x,y,s=sizes,alpha=0.5)
```

```python
#random colors for random points 
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='nipy_spectral')
plt.colorbar()
plt.show()
plt.savefig('Test2.png')
```
