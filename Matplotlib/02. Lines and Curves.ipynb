# Lines and Curves
We will be plotting lines and curves with given points and equations and for that we will use `plot()` method, which plots a line or curve btwn (x1, y1) and (x2, y2).  
The syntax is `.plot([x1,x2],[x2,y2])`

#### Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Drawing a line for given points

```python
#Line 1
#draw a line from (3,1) to (5,6) color='green'
plt.plot([3,5],[1,6],color='green')

#Line 2
#draw a line from (0,0) to (5,5)
plt.plot([0,5],[0,5],color='red')

#give labels to the axis
plt.xlabel("X")
plt.ylabel("Y")

#giving the title to the graph

plt.title("Plot arbitary to lines")
plt.show() #to show off course
```

####  Drawing a curve of the function

```python
x=np.array([1,2,3,4])#X-axis points

#using function y=x^2
y=x*x #Y axis points

plt.plot(x,y,c='b')#plot a chart

#give labels to the axis
plt.xlabel("X")

plt.ylabel("Y")
#giving the title to the graph
plt.title("y=x^2")

plt.show() #to show
#we are taking the x as descrete therefore the curve is not smooth
```

```python
#Using function y=x^3
x=np.array([1,2,3,4])#X-axis points
y=x**3 #Y axis points
plt.plot(x,y,c='b')#plot a chart
#give labels to the axis
plt.xlabel("X")
plt.ylabel("Y")
#giving the title to the graph
plt.title("y=x^3")
plt.show() #to show offcourse
#we are taking the x as descrete therefore the curve is not smooth
```

One can use `.arange` instead of `array`.

```python
x=np.arange(1,4)#X-axis points
y=x*x #Y axis points
plt.plot(x,y,c='b')#plot a chart
#give labels to the axis
plt.xlabel("X")
plt.ylabel("Y")
#giving the title to the graph
plt.title("y=x^2")
plt.show() #to show offcourse
#we are taking the x as descrete therefore the curve is not smooth
```

For smooth curve you need to lake lot of points in smaller area, that can be done by taking lot of points or increasing the number of steps in `.arange`.

```python
#Example
x=np.arange(0,20)
y=x**3
plt.plot(x,y,c='g')
plt.xlabel("X axis")
plt.ylabel("Y axis")
#dont take smaller steps in order to get the smooth curve your laptop can't process that
#if you are really eager to do that `x=np.arange(0,20,0.000001)` make this one change
```

#### Adding Marker

```python
x=np.arange(0,20)
y=3*x^2+5
plt.plot(x,y,'b*',linestyle='dashed',linewidth=2,markersize='15')
```

#### Multiple Charts in Same Container

```python
#Chart 1
x=np.array([1,2,3,4,5])
y=2*x+5
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('Line 1')

#the figure() function helps us to create new figure with same container
plt.figure()

#Chart 1
x1=[2,4,6,8]
y1=[3,5,7,9]
plt.plot(x1,y1,'-.')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title('Line 2')
```
