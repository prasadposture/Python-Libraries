# Histogram  
A histogram represents the frequencies of values of a variable bucketed into ranges.
Histogram is similar to bar chat but the difference is it groups the values into continuous ranges.
Each bar in histogram represents the height of the number of values present in that range.

#### Importing Libraries

```python
from matplotlib import pyplot as plt
import numpy as np
```

#### Creating a histogram
We use `hist()` function provided by `matplotlib`.

```python
#Example
marks=np.array([22,87,5,43,56,44,67,73,55,54,11,20,51,5,79,31,27])
plt.hist(marks)
plt.show()
```

#### Defining Bin Size
Defininig a bin size helps us divide the numeric data into equal size such that bin size can achieved.

```python
plt.hist(marks,bins=20) #Creates 20 bins
plt.show()
```

We can also define bin size based on the ranges we want. This helps in achieving bins of varying widths.

```python
#Example : Suppose you want 4 Bins, 0 - 25, 26-50, 51-75, 76-100
plt.hist(marks,bins=[0,25,50,75,100])
plt.show()
```

#### Adding title and labels

```python
plt.hist(marks,bins=[0,25,50,75,100])
plt.title("Histogram of Results") #title of the plot
plt.xlabel("Marks") #X-axis label
plt.ylabel("Number of students") #Y-axis label
plt.show()
```

**Note :** <br>
The problem is that there is no proper boundary of the histogram bars and we are not able to distinguish between them. The color is same of all the bars and also we want to adjust the alignment. <br> 
We can overcome this difficulty by making following changes in bar:<br>
1. Alignment- mid, left, right, bottom <br>
2. color of the bars<br>
1. Edgecolor of the histogram bars<br>

#### Adding Color, Edgecolor and Alignment

```python
plt.hist(marks,bins=[0,25,50,75,100],align='right',color='purple',edgecolor='black')
plt.title("Histogram of Results")
plt.xlabel("Marks")
plt.ylabel("Number of students")
plt.show() #now the histogram bars are clearly visible.
```

#### Creating Histogram with Pandas

```python
#Importing Pandas
import pandas as pd
```

```python
#Creating dataframe
df=pd.DataFrame(marks)
```

```python
#Plotting with Pandas
df.hist() #using hist() method on dataframe directly gives us the histogram
plt.title("Histogram of Results")
plt.xlabel("Marks")
plt.ylabel("Number of students")
plt.show()
```

Histogram is a particular case of the bar diagram when the bars are continuous. We will see an example supporting this statement.

```python
#Example : Without speicifying the range
blood_suagr=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_suagr,rwidth=0.8)
plt.title("Blood Sugar Chart")
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.show()
#it comes as bar diagram since the range isnt specified
```

```python
#Example : With specifing the range
plt.title("Blood Sugar Chart")
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.hist(blood_suagr,bins=[80,100,125,150],rwidth=0.95,color='g')
plt.show()
#Gives a histogram for given ranges
```

#### Different bar colors for multiple data samples and the legend

```python
#Titles and Labels
plt.title("Blood Sugar Chart")
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")

#Data Samples
blood_suagr_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]

#Histogram Plot
plt.hist([blood_suagr_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=["orange","green"],label=["men","women"])
plt.legend()
plt.show()
```

#### Changing the orientation

```python
#Titles and Labels
plt.title("Blood Sugar Chart")
plt.xlabel("Number of Patients")
plt.ylabel("Sugar Levels")

#Plot
plt.hist([blood_suagr_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=["orange","green"],label=["men","women"],orientation="horizontal")
plt.legend()
plt.show()
```
