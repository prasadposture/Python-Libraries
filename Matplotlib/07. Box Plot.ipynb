# Box Plot
Boxplots are a measure of how well distributed is the data in a data set. It divides the
data set into three quartiles.

### Graph of box
This graph represents the minimum, maximum, median, first quartile and third quartile
in the data set.

```python
#Importing Dependencies
import matplotlib.pyplot as plt
```

### Creating a boxplot

```python
#values
values=[84,77,20,40,67,62,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52]
#plotting boxplot
plt.boxplot(values)
plt.show()
```

### Filling up the boxes with color

```python
#values
value1=[84,77,20,40,67,62,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52]
value2=[62,5,91,25,35,52,96,90,95,96,99,3,90,95,34,55,27,100,15,16,16,75,65,31,25,52]
value3=[23,89,12,78,72,89,25,69,68,86,19,48,15,16,16,65,31,25,52]
value4=[59,73,16,81,88,98,10,87,29,72,16,23,72,116,23,72,88,78,99,75,30]

#list values
box_plot_data=[value1,value2,value3,value4]

#labels
labels=['subject1', 'subject2','subject3','subject4']

#plotting boxplot
plt.boxplot(box_plot_data, labels=labels,patch_artist=True)

#titles and labels to the axis
plt.title("Boxplot of Marks")
plt.ylabel("Marks of the students")
plt.xlabel("Name of the subject")

plt.show()
```

```python
#values
value1=[84,77,20,40,67,62,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52]
value2=[62,5,91,25,35,52,96,90,95,96,99,3,90,95,34,55,27,100,15,16,16,75,65,31,25,52]
value3=[23,89,12,78,72,89,25,69,68,86,19,48,15,16,16,65,31,25,52]
value4=[59,73,16,81,88,98,10,87,29,72,16,23,72,116,23,72,88,78,99,75,30]

#list values
box_plot_data=[value1,value2,value3,value4]

#labels
labels=['subject1', 'subject2','subject3','subject4']

#plotting boxplot
plt.boxplot(box_plot_data, labels=labels, patch_artist=True, notch=True)

#titles and labels to the axis
plt.title("Boxplot of Marks")
plt.ylabel("Marks of the students")
plt.xlabel("Name of the subject")

plt.show()
```

### Boxplot with Pandas Data Frame

```python
#Importing Dependencies
import pandas as pd
import numpy as np
```

```python
#Reading the data
df=pd.read_csv('water_potability.csv')
#Dropping the response variable
df.drop(['Potability'],axis=1, inplace=True)
#Dropping null values
df.dropna(axis=0,inplace=True)
df.head()
```

```python
#adjusting the range of the values in the column from 0 to 1
for x in list(df.columns):
    df[x]=df[x]/max(df[x])
df.head()
```

### Plotting boxplot for the above dataframe

```python
#plotting boxplot
plt.boxplot(df, #dataframe used
           labels=list(df.columns), #adding labels
           patch_artist=True #filling up the boxes color
           )
plt.title('Boxplot for water potability') #title of the boxplot
plt.xlabel('Affecting Factors')
plt.ylabel('Impact Measure')
plt.xticks(rotation=90) #changing the orientation of the xticks
plt.show()
```
