# Scatter Plot

#### Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np #NumPy helps matplotlib
```

#### Creating scatter plot
We use `scatter()` method from matplotlib for plotting scatterplot.

```python
#Example
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)
plt.show() #for showing the scatter plot
```

#### Comparing two scatter plots

```python
#day ONE, the age and speed of 13 cars
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)

#day TWO, the age and speed of 13 cars
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y)

plt.show()
```

#### Changing the colors

```python
#day ONE, the age and speed of 13 cars
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y,color='hotpink') #assgning color using name of the color

#day TWO, the age and speed of 13 cars
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y,color='#88c999') #assgning color using ASKII character

plt.show()
```

#### Labelling the axis and giving a title 

```python
#Example
x=np.arange(0,10) #using range
y=np.arange(11,21)
plt.scatter(x,y,c='green')

#labels
plt.xlabel("X axis")
plt.ylabel("Y axis")

#title of the graph
plt.title("Graph in 2D")

#saving the graph
plt.savefig('Test1.png')
```

#### Giving each point a different color
This can be used for representing 3D plots over 2D.

```python
#Example
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors=np.array(['red','green','blue','yellow','orange','cyan','beige','skyblue','pink','brown','black','gray','purple'])
plt.scatter(x,y,c=colors) #assigning colors to color vector
```

#### Giving each point a different size
This can be used for representing 3D plots over 2D.

```python
#Example
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,77])
plt.scatter(x,y,s=sizes)
```

#### Adjusting the transperancy of the colors

```python
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,77])
plt.scatter(x,y,s=sizes,alpha=0.5)
```

#### Random colors and sizes for Random points 

```python
#Example
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='nipy_spectral')
plt.colorbar()
plt.show()
plt.savefig('Test2.png')
```

#### Plotting with grid
Useful for negative numbers

```python
x=np.array([-2,-1,0,1,2])
y=np.array([-5.35,-5,-4,-3,2.25])
plt.scatter(x,y,c='green')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph in 2D")
plt.grid()
plt.show()
plt.savefig('assignment.png')
```
