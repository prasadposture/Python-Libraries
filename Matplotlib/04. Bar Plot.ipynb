# Bar Plot

Bar Plot is used to show relationship between catageory and some numeric value.  
The general syntax is `plt.bar(x,height,width,color,bottom,align)`

#### Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Creating a Bar Plot

```python
#on x axis
x=np.array(['A','B','C','D'])

#on y axis
y=np.array([2,8,1,10])

#plotting
plt.bar(x,y,color='r')

#giving labels and titles
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar Graph")

#showing the plot
plt.show()
```

#### Plotting Two Different Bars

```python
#for first plot
x1=[2,8,10]#characteristic
y1=[11,16,9]#value

#for second plot
x2=[3,9,11]
y2=[6,15,7]

#plot 1
plt.bar(x1,y1)

#plot 2
plt.bar(x2,y2,color='g')

#giving title and labelling the axis
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar Graph")
plt.show()
```

#### Adjusting the width of the bar chart

```python
x=['APPLE','BANANA']
y=[400,350]
plt.bar(x,y,color='r')
plt.show()
```

Width of the bar is so much doesnt look perfect one can change it as well
by defining the width

```python
x=['APPLE','BANANA']
y=[400,350]
plt.bar(x,y,width=0.2,color='r') #adjusting the width
plt.show()
```

#### Adding New Colors
Check the names and ASKII codes of the colors here : https://htmlcolorcodes.com/color-names/

```python
#Example
x=np.array(['A','B','C','D'])
y=np.array([2,8,1,10])
plt.bar(x,y,color='darkcyan')#adding new colors
#giving title and labelling the axis
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar Graph")
plt.show()
```

#### Horizontal Bar

```python
x=np.array(['Rice','Bread','Tea','Sugar'])
y=np.array([120,14,100,75])
plt.barh(x,y,color="RebeccaPurple",height=0.5)
plt.show()
```

### Working with Pandas

```python
#import library
import pandas as pd
```

```python
#create data frame
df=pd.DataFrame.from_dict(
{
    'Year':[2016,2017,2018,2019,2020,2021],
    'Men':[300,320,310,305,325,330],
    'Women':[350,325,300,305,380,355],
    'Total':[650,645,610,610,705,685]
}
)
```

```python
#first five entries
df.head()
```

#### Creating a bar chart using dataframe

```python
#Create a bar diagram
#giving different colors for different bars
colors=['Crimson','darkcyan','Olive','black','Khaki','LightSeaGreen']

#Using attributes from the dataframe for plotting
plt.bar(x=df['Year'],height=df['Total'],color=colors)

#title and labels
plt.title("Sales over time",fontsize=20)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.show()
```

#### Double Bar Chart
Double bar chart helps us in comparing different values for different attributes.

```python
#Example : Plotting double bar chart of men and women 

#plot 1
plt.bar(x=df['Year'],height=df['Men'])

#plot2
plt.bar(x=df['Year'],height=df['Women'])

#title and labels
plt.title("Sales over time",fontsize=20)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.show()
```

But this seems an inconvinient way since most of short bars are hidden behind taller bars, also we can't plot legend for the same for distinguishing them. 

```python
#double bar chart : keeping bars side by side
width=0.4
colors=['Crimson','darkcyan','Olive','black','Khaki','LightSeaGreen']

#plot 1
plt.bar(x=df['Year'],height=df['Men'],width=width,label='Men')

#plot2
plt.bar(x=df['Year']+width,height=df['Women'],width=width,label='Women')

#title, labels and legend
plt.title("Sales over time",fontsize=20)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.legend()
plt.show()
```

We can split the bar of `Total` into it's attributes for plotting `Men` and `Women` on the single bar

```python
#Men and Women on the single bar
plt.bar(x=df['Year'],height=df['Total'],color='r',label='Women')
plt.bar(x=df['Year'],height=df['Men'],color='b',label='Men')
plt.title("Sales over time",fontsize=20)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.legend()
plt.show()
```
